# 1. Email Validator (String Filtering + Loops)
üíº Real-world use: Used in signup forms or data cleaning tasks.

Task:

Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù.

Example:

Input: hello@inceptionbd.com

Output: Valid Email

```python
email = input("Please enter an email address: ")

at = False
space = False

for character in email:
    if character == '@':
        at = True
    if character == ' ':
        space = True


ending1 = email.endswith(".com")
ending2 = email.endswith(".org")


if at and (not space) and (ending1 or ending2):
    print("\nValid Email")
else:
    print("\nInvalid Email")
```

# 2. Count Word Frequency in Customer Reviews
üíº Real-world use: Analyzing reviews or comments in text analytics.

Task:

Given a review text, count how many times each word appears (ignore case).

Example:

Input: "This product is good and this service is excellent"

Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
review = input("Please write a review: ")

review = review.lower()
words = review.split()

flag = {}

for i in words:
    if i in flag:
        flag[i] += 1
    else:
        flag[i] = 1

print()
print(flag)
```

# 3. Password Strength Checker

üíº Real-world use: Used in user authentication systems.

Task:

Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one
special character.

Example:

Input: "Hello@123"

Output: Strong Password

```python
password = input("Please enter a password: ")

upper = False
lower = False
digit = False
special = False

sp_char = "!@#$%^&*()-_=+[]{},.<>?/|"

for i in password:
    if i.isupper():
        upper = True
    elif i.islower():
        lower = True
    elif i.isdigit():
        digit = True
    elif i in sp_char:
        special = True


if password == upper and lower and digit and special:
    print("\nStrong Password")
else:
    print("\nWeak Password")
```

# 5. SMS Word Counter

üíº Real-world use: Used in messaging apps or social media post limit check.

Task:

Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).

Example:

Input: "This is a demo message."

Output: Words = 5, Characters = 23, Status = Within Limit

```python
messege = input("Please enter message: ")

characters = len(messege)

words = len(messege.split())

if characters > 160:
    status = "Exceeded Limit"
else:
    status = "Within Limit"

print("\nWords =",words,",", "Characters =", characters,",","Status =" , status)
```

# 7. Detect Suspicious Words in Text (Moderation)

üíº Real-world use: Used in chat moderation or email filtering.

Task:

If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.

Example:

Input: "This is a scam offer"

Output: Warning: Suspicious content detected

```python
banned_words_list = ["spam", "scam", "fake"]

ban = input("Please enter your text: ").lower()

warning = False

for i in banned_words_list:
    if i in ban:
        warning = True
        break

if warning:
    print("\nWarning: Suspicious content detected")
else:
    print("\nNo suspicious content found")
```

# 9. Count Hashtags and Mentions (Social Media Analytics)

üíº Real-world use: Used in analyzing social media posts.

Task:

Count how many hashtags (#) and mentions (@) appear in a post.

Example:

Input: "Check out #AIWorkshop and follow @inceptionbd"

Output: Hashtags = 1, Mentions = 1

```python
post = input("Please enter your social media post: ")

hashtag_count = 0
mention_count = 0

words = post.split()

for i in words:
    if i.startswith('#'):
        hashtag_count += 1
    elif i.startswith('@'):
        mention_count += 1

print(f"\nHashtags = {hashtag_count}, Mentions = {mention_count}")
```

# 10. Word Reversal for Data Encryption

üíº Real-world use: Simple text obfuscation or playful encoding.

Task:

Reverse each word in a sentence while keeping word order the same.

Example:

Input: "AI is fun"

Output: "IA si nuf"

```python
sentence = input("Please enter a sentence: ")

w = sentence.split()
rev_words = []

for i in w:
    rev_words.append(i[::-1])

reversed_sentence = " ".join(rev_words)

print()
print(reversed_sentence)
```
