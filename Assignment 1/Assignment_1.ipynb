# 1. Temperature Check

Problem: Write a program that takes the current temperature as input and prints:

● "It's too cold!" if below 10°C

● "It's cool outside" if between 10°C and 25°C

● "It's hot outside!" if above 25°C

```python
temperature = float(input("Please enter current temperature : "))

if temperature <10:
    print("\nIt’s too cold!")

elif temperature >=10 and temperature <=25  :
    print("\nIt’s too cool outside!")

elif temperature >25  :
    print("\nIt’s hot outside!")


```

# 2. ATM Withdrawal System

Problem: Ask the user for their account balance and the amount they want to withdraw.

● If the withdrawal amount is greater than the balance → print "Insufficient Balance"

● If the withdrawal amount is less than or equal to balance → print "Transaction Successful" and show the remaining balance.

```python
account_balance = int(input("Please enter your account balance : "))
withdrawal_amount = int(input("Please enter the amount you want to withdraw : "))
remain_balance = account_balance- withdrawal_amount


if withdrawal_amount >  account_balance:
    print("\nInsuficient Balance!")

elif withdrawal_amount <= account_balance:
    print("\nTransaction Successfull!")
    print("\nYour Remaining Balance is = ", remain_balance )

```

# 3. Student Grade Calculator

Problem: Input marks (0–100) and print grade based on:

● 90–100 → A+

● 80–89 → A

● 70–79 → B

● 60–69 → C

● Below 60 → Fail

```python
mark = int(input("Enter your marks(0-100) : "))

if mark >= 90 or mark == 100:
    print("\nA+")
elif mark >= 80 or mark == 89:
    print("\nA")
elif mark >= 70 or mark == 79:
    print("\nB")
elif mark >= 60 or mark == 69:
    print("\nC")
elif mark < 60:
    print("\nFail")
```

# 5. Restaurant Bill With Discount

Problem: Take total bill amount from the user:

● If the bill is above 1000 → apply 10% discount

● Otherwise → no discount Then print the final bill amount.

```python
total_bill = int(input("Please enter your total bill : "))
new_total_bill = int((total_bill * 10) / 100)

if total_bill > 1000:
    print("\nFinal Bill is: ", total_bill - new_total_bill)

else:
    print("\nNo discount. Your Final bill is : ",total_bill )
```

# 7. Traffic Light System

Problem: Ask the user to input a color (red, yellow, or green). Print:

● "Stop" for red

● "Ready to go" for yellow

● "Go" for green

● "Invalid color" otherwise

```python
colour = input("Please enter colour (red,yellow or green) : ")

r = "red"
y = "yellow"
g = "green"

if colour == r:
    print("\nStop")
elif colour == y:
    print("\nReady to go")
elif colour == g:
    print("\nGo")
    
else:
    print("\nInvalid Colour")
```

# 8. Leap Year Checker

Problem: Take a year as input and check:

● If it is divisible by 400 → Leap Year

● Else if divisible by 4 but not by 100 → Leap Year

● Otherwise → Not a Leap Year

```python
year = int(input("Please enter the year : "))

l1 = int(year%400)
l2 = int(year%4)
l3 = int(year%100)


if l1 == 0:
    print("\nLeap Year")

elif l2 == 0 and l3 != 0:
    print("\nLeap Year")

else:
    print("\nNot a Leap Year")

```

# 10. Electricity Bill Calculator

Problem: Input number of units consumed and calculate bill:

● Up to 100 units → 5 BDT per unit

● 101–200 units → 7 BDT per unit

● Above 200 units → 10 BDT per unit Print total payable amount.

```python
unit = int(input("Please enter the number of units : "))

p1 = int(unit*5)
p2 = int(unit*7)
p3 = int(unit*10)

if unit <= 100:
    print("\nTotal payable amount :", p1)

elif unit > 100 and unit <= 200:
    print("\nTotal payable amount :", p2)

elif unit > 200:
    print("\nTotal payable amount :", p3)
```
